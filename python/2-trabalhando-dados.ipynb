```python
import pandas as pd
import numpy as np 
# Loading the labeled decisions
FILE_DECISION_CSV = "data/2019-04-07_20h10m27s-decisions-no-repetitions.csv"
data = pd.read_csv( FILE_DECISION_CSV, sep='<=>',header=0,engine='python')
data.head()
```

# Separação das colunas mais importantes e calculo do tamanho da ementa

```python
tamanho = data['ementa_text'].apply( lambda x : len(x) if x else 0)
```

```python
data_importante = pd.concat( [data['orgao_julgador'], data['process_type'], data['decision_text'], data['decision_label'], data['ementa_text'] ], axis=1)
data_importante['tamanho_ementa'] = tamanho
data_importante.head()
```

# Verificar dado sem ementa

```python
data_importante[data_importante['tamanho_ementa'] < 100]
```

# Tratar dados 

```python
data_tratado = data_importante[data_importante['tamanho_ementa'] > 100]
```

# Exportar dados

```python
data_tratado.reset_index()
data_tratado.to_csv('data/dados_tratados.csv', sep = '~')
```
