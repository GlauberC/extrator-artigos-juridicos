# Remover stopwords, pontuação e acentuação

```python
def remover_stopwords_pontuacao_acentuacao(data):
    # Recebe um dataframe e retorna uma lista dos dados sem stopwords e sem pontuação e acentuacao
    pontuacao = list()

    for p in punctuation:
        pontuacao.append(p)

    stopwords = nltk.corpus.stopwords.words("portuguese")
    frases_processadas = list()

    pontuacao_stopwords = stopwords + pontuacao

    frases_processadas = list()
    for d in data:
        nova_frase = list()
        token_pontuacao = tokenize.WordPunctTokenizer()
        palavras = token_pontuacao.tokenize(unidecode.unidecode(d.lower()))

        for palavra in palavras:
            if palavra not in pontuacao_stopwords:
                nova_frase.append(palavra)

        frases_processadas.append(' '.join(nova_frase))

    return frases_processadas
```

# Stemmização

```python
def stemmizacao(data):
    # Recebe um dataframe e retorna uma lista dos dados contendo apenas o radical das palavras

    stemmer = nltk.RSLPStemmer()

    frases_processadas = list()
    for d in data:
        nova_frase = list()
        token_espaco = tokenize.WhitespaceTokenizer()
        palavras = token_espaco.tokenize(d)

        for palavra in palavras:
            nova_frase.append(stemmer.stem(palavra))

        frases_processadas.append(' '.join(nova_frase))

    return frases_processadas
```

```python
df['tratamento2'] = stemmizacao(df['tratamento1'])
```

# Núvem de palavras

```python
def nuvem_de_palavras(dataframe):
    # Recebe uma coluna de um dataframe com frases e exibe uma núvem de palavras
    
    %matplotlib inline
    todas_frases = ' '.join([texto for texto in dataframe])
    nuvem_palavras = WordCloud(width = 1200, 
                               height = 800,
                               max_font_size = 110,
                               collocations = False).generate(todas_frases)
    plt.axis("off")
    plt.figure(figsize = (20,15))
    plt.imshow(nuvem_palavras, interpolation="bilinear")
```

# Frequencia das palavras

## Frequencia - Tratamento 1

```python
def frequencia_palavras(dataframe):
    frase = ' '.join(dataframe)

    token_espaco = tokenize.WhitespaceTokenizer()
    token_frase = token_espaco.tokenize(frase)

    frequencia = nltk.FreqDist(token_frase)
    df_frequencia = pd.DataFrame({"Palavra": list(frequencia.keys()),
                                 "Frequencia": list(frequencia.values())})
    return df_frequencia

```

```python
freq = frequencia_palavras( df['tratamento1'] )
freq.head(10)
```

## Frequencia - Tratamento 2

```python
freq = frequencia_palavras( df['tratamento2'] )
freq.head(10)
```

```python
def barplot_palavras(df, nome_coluna_Y, nome_coluna_X,  quantidade_de_palavras = 15):
    
    freq = frequencia_palavras( df )
    
    %matplotlib inline
    plt.figure(figsize=(20,10))
    ax = sns.barplot(data = freq.nlargest(columns = nome_coluna_Y, n = quantidade_de_palavras), x = nome_coluna_X, y = nome_coluna_Y, color = "green" )
    ax.set(ylabel = "Contagem")
```

## Barplot - Tratamento 1

```python
barplot_palavras(df['tratamento1'], "Frequencia", "Palavra")
```

# TfIdf

```python
def calcula_Tfidf(dataframe, juntar = False):
    # calcula o TF-IDF e retorna um dataframe, se juntar = True: ele considera todos os dados como um dado só (TFIDF total da base dados)
    
    tfidf = TfidfVectorizer()
    
    if juntar:
        todas_frases = ' '.join([texto for texto in dataframe])
        caracteristicas = tfidf.fit_transform([todas_frases])
    else:
        caracteristicas = tfidf.fit_transform(dataframe)
    
    novo_df = pd.DataFrame(
        caracteristicas.todense(),
        columns = tfidf.get_feature_names()
    )
    return novo_df
    
```

```python
def seleciona_faixa_tf_idf(dataframe, minima =  0.000001, maxima = 100):
    # recebe o dataframe calculado com o tf-idf e retorna o dataframe contendo as palavras contida no range escolhido
    
    novo_df = dataframe[(dataframe < maxima) & (dataframe > minima) ]
    return novo_df.dropna(axis = 1)
```

# Eliminar palavras fora do alcance

```python
def eliminar_palavras_fora_faixa(data, minima =  0.000001, maxima = 100, remove_digit = False):
    # Elimina as palavas que não estão na faixa do tfidf. caso remove_digit = true: não serão aceitas palavras com digitos
    
    df_tf_idf = calcula_Tfidf(data, True)
    df_faixa = seleciona_faixa_tf_idf(df_tf_idf, minima, maxima)
    

    
    palavras_faixa = list(df_faixa.columns)
    if (remove_digit):
        new_palavras_faixa = list(df_faixa.columns)
        for palavra_faixa in palavras_faixa:
            if palavra_faixa.isdigit():
                new_palavras_faixa.remove(palavra_faixa)
    
    palavras_faixa = new_palavras_faixa
    
    frases_processadas = list()
    for d in data:
        nova_frase = list()
        token_espaco = tokenize.WhitespaceTokenizer()
        palavras = token_espaco.tokenize(d)

        for palavra in palavras:
            if palavra in palavras_faixa:
                nova_frase.append(palavra)

        if(len(nova_frase) < 1):
            frases_processadas.append('')
        else:
            frases_processadas.append(' '.join(nova_frase))
    
    return frases_processadas
    
```

```python
def vetorizar_palavras(dataframe, numero_de_palavras):
    # Recebe uma coluna de um dataframe de frases e retorna um dataframe vetorizado
    
    modelo = CountVectorizer(max_features = numero_de_palavras)
    bag = modelo.fit_transform(dataframe)
    new_df = pd.SparseDataFrame(bag,
                     columns = modelo.get_feature_names())
    new_df.fillna(0.0, inplace = True)
    return new_df
```
